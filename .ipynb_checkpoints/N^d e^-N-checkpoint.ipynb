**Task:** Plot $N^{d}e^{-N}$ for different $d$ values. Study the values of $N$ to obtain $N^{d}e^{-N}=0.1$ and $N^{d}e^{-N}=0.01$.

**Solution**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
# Define the values for d and N
d_values = [5, 10, 15, 20, 25, 30]
N = list(range(1, 201))
```

```python
import pandas as pd
f, ax = plt.subplots(figsize=(7, 7))
for d in d_values:
    values = [(n**d)*(math.exp(-n)) for n in N]
    df = pd.DataFrame(list(zip(N, values)),
                      columns =['N', 'values'])   
    ax.set(yscale="log")
    plt.plot(df.iloc[:, 0], df.iloc[:, 1])    

plt.axhline(10**-4, color = 'indigo', linestyle = '--')
plt.axhline(10**-7, color = 'm', linestyle = ':')
plt.ylabel(r'$N^{d}e^{-N}$')
plt.xlabel(r'$N$')
plt.ylim([10**-10,10**10])
plt.title(r'How much data do we need for a certain $d$?')
ax.legend(['d=5', 'd=10', 'd=15', 'd=20', 'd=25', 'd=30', r'delta = $10^{-4}$', r'delta = $10^{-7}$'])
plt.show()
```

We note that:
- If $d=5$, then for $10^{-10}$ we will need $\geq 50$ data samples.
- If $d=10$, then for $10^{-10}$ we will need $\geq 75$ data samples.
- If $d=15$, then for $10^{-10}$ we will need $\geq 100$ data samples.
- If $d=20$, then for $10^{-10}$ we will need $\geq 125$ data samples.
- If $d=25$, then for $10^{-10}$ we will need $\geq 150$ data samples.
- If $d=30$, then for $10^{-10}$ we will need $\geq 200$ data samples.

We can see that for almost all cases $N\geq 7d$, however a more accurate bound will be $N\approx 10d$.
