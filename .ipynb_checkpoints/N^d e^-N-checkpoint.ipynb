**Task:** Plot $N^{d}e^{-N}$ for different $d$ values. Study the values of $N$ to obtain $N^{d}e^{-N}=0.1$ and $N^{d}e^{-N}=0.01$.

**Solution**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
# Define the values for d and N
d_values = [5, 10, 15, 20]
N = list(range(1, 201))
```

```python
import pandas as pd
f, ax = plt.subplots(figsize=(7, 7))
for d in d_values:
    values = [(n**d)*(math.exp(-n)) for n in N]
    df = pd.DataFrame(list(zip(N, values)),
                      columns =['N', 'values'])   
    ax.set(xscale="log")
    plt.plot(df.iloc[:, 1], df.iloc[:, 0])    

plt.axvline(0.1, color = 'indigo', linestyle = '--')
plt.axvline(0.01, color = 'm', linestyle = ':')
plt.xlabel(r'$N^{d}e^{-N}$')
plt.ylabel(r'$N$')
plt.title(r'How much data do we need for a certain $d$?')
ax.legend(['d=5', 'd=10', 'd=15', 'd=20', 'eps = 0.1', 'eps = 0.01'])
plt.show()
```

We note that:
- If $d=5$, then $25 \ge N_{0.01}>N_{0.1}$.
- If $d=10$, then $50 \ge N_{0.01}>N_{0.1} \ge 25$.
- If $d=15$, then $75 \ge N_{0.01}>N_{0.1}\ge 50$.
- If $d=20$, then $100 \ge N_{0.01}>N_{0.1} \ge 75$.

We can see that for these cases $N\approx 5d$, however a more accurate bound will be $N\approx 10d$.
